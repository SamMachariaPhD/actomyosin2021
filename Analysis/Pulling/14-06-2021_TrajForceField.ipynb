```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
cm = 1/2.54
import glob
```

```python
f0_0l = glob.glob('data/tip/TipXYF0.0*.txt')
f0_25l = glob.glob('data/tip/TipXYF0.25*.txt')
f0_5l = glob.glob('data/tip/TipXYF0.5*.txt')
f1_0l = glob.glob('data/tip/TipXYF1.0**.txt')
f1_5l = glob.glob('data/tip/TipXYF1.5**.txt')
f_0_25l = glob.glob('data/tip/TipXYF-0.25**.txt')
f_0_5l = glob.glob('data/tip/TipXYF-0.5**.txt')
f_1_0l = glob.glob('data/tip/TipXYF-1.0**.txt')
```

```python
f0_0 = []
for i in f0_0l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True)
    f0_0.append(_)

f0_25 = []
for i in f0_25l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True)
    f0_25.append(_)

f0_5 = []
for i in f0_5l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True)
    f0_5.append(_)
    
f1_0 = []
for j in f1_0l:
    _ = pd.read_csv(j, names=['t','x','y'], delim_whitespace=True)
    f1_0.append(_)    
    
f1_5 = []
for k in f1_5l:
    _ = pd.read_csv(k, names=['t','x','y'], delim_whitespace=True)
    f1_5.append(_)
    
f_0_5 = []
for l in f_0_5l:
    _ = pd.read_csv(l, names=['t','x','y'], delim_whitespace=True)
    f_0_5.append(_)
    
f_0_25 = []
for i in f_0_25l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True)
    f_0_25.append(_)
    
f_1_0 = []
for m in f_1_0l:
    _ = pd.read_csv(m, names=['t','x','y'], delim_whitespace=True)
    f_1_0.append(_)
```

```python
fig, ax = plt.subplots(2,3, figsize=(12,6), sharex=True, sharey=True, constrained_layout=True)
fig.subplots_adjust(hspace=-0.5)
fig.subplots_adjust(wspace=0.05)
#plt.subplot_tool()

for i in range(4):
    ax[0,2].plot(f_1_0[i]['x'],f_1_0[i]['y'],lw=1)
    ax[0,2].set_aspect('equal')
    ax[0,2].minorticks_on()
    ax[0,2].tick_params(which='both',direction='in', top=True, right=True)
    #ax[0,2].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[0,0].set_ylabel('Y ($\mu m$)', fontsize=16)
    ax[0,2].set_title('$F = -1.0\ pN/\mu m$', fontsize=16)
    
for i in range(4):
    ax[0,1].plot(f_0_5[i]['x'],f_0_5[i]['y'],lw=1)
    ax[0,1].set_aspect('equal')
    ax[0,1].minorticks_on()
    ax[0,1].tick_params(which='both', direction='in', top=True, right=True)
    #ax[0,1].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[0,1].set_title('$F = -0.5\ pN/\mu m$', fontsize=16)
    
for i in range(4):
    ax[0,0].plot(f_0_25[i]['x'],f_0_25[i]['y'],lw=1)
    ax[0,0].set_aspect('equal')
    ax[0,0].minorticks_on()
    ax[0,0].tick_params(which='both', direction='in', top=True, right=True)
    #ax[0,0].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[0,0].set_title('$F = -0.25\ pN/\mu m$', fontsize=16)
        
# for i in range(4):
#     ax[1,1].plot(f0_0[i]['x'],f0_0[i]['y'],lw=1)
#     ax[1,1].set_aspect('equal')
#     ax[1,1].minorticks_on()
#     ax[1,1].tick_params(which='both', direction='in', top=True, right=True)
#     #ax[1,1].set_xlabel('X ($\mu m$)', fontsize=16)
#     ax[1,1].set_title('$F = 0.0\ pN/\mu m$', fontsize=16)
    
for i in range(4):
    ax[1,0].plot(f0_25[i]['x'],f0_25[i]['y'],lw=1)
    ax[1,0].set_aspect('equal')
    ax[1,0].minorticks_on()
    ax[1,0].tick_params(which='both', direction='in', top=True, right=True)
    #ax[1,0].set_ylabel('Y ($\mu m$)', fontsize=16)
    ax[1,0].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[1,0].set_title('$F = 0.25\ pN/\mu m$', fontsize=16)
    
for i in range(4):
    ax[1,1].plot(f0_5[i]['x'],f0_5[i]['y'],lw=1)
    ax[1,1].set_aspect('equal')
    ax[1,1].minorticks_on()
    ax[1,1].tick_params(which='both', direction='in', top=True, right=True)
    ax[1,1].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[1,1].set_title('$F = 0.5\ pN/\mu m$', fontsize=16)
    
for i in range(4):
    ax[1,2].plot(f1_0[i]['x'],f1_0[i]['y'],lw=1)
    ax[1,2].set_aspect('equal')
    ax[1,2].minorticks_on()
    ax[1,2].tick_params(which='both', direction='in',top=True, right=True)
    ax[1,2].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[1,0].set_ylabel('Y ($\mu m$)', fontsize=16)
    ax[1,2].set_title('$F = 1.0\ pN/\mu m$', fontsize=16)
    
# for i in range(4):
#     ax[1,3].plot(f1_5[i]['x'],f1_5[i]['y'],lw=1)
#     ax[1,3].set_aspect('equal')
#     ax[1,3].minorticks_on()
#     ax[1,3].tick_params(which='both', direction='in', top=True, right=True)
#     ax[1,3].set_xlabel('X ($\mu m$)', fontsize=16)
#     ax[1,3].set_title('$F = 1.5\ pN/\mu m$', fontsize=16)

#plt.tight_layout()

#plt.savefig('fig/pulling_fieldF-1to1.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
