```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
cm = 1/2.54
import glob
```

```python
f0_0l = glob.glob('data/tip/TipXYF0.0*.txt')
f1_0l = glob.glob('data/tip/TipXYF1.0**.txt')
f1_5l = glob.glob('data/tip/TipXYF1.5**.txt')
f_0_5l = glob.glob('data/tip/TipXYF-0.5**.txt')
f_1_0l = glob.glob('data/tip/TipXYF-1.0**.txt')
```

```python
f0_0 = []
for i in f0_0l:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True)
    f0_0.append(_)
    
f1_0 = []
for j in f1_0l:
    _ = pd.read_csv(j, names=['t','x','y'], delim_whitespace=True)
    f1_0.append(_)
    
    
f1_5 = []
for k in f1_5l:
    _ = pd.read_csv(k, names=['t','x','y'], delim_whitespace=True)
    f1_5.append(_)
    
f_0_5 = []
for l in f_0_5l:
    _ = pd.read_csv(l, names=['t','x','y'], delim_whitespace=True)
    f_0_5.append(_)
    
f_1_0 = []
for m in f_1_0l:
    _ = pd.read_csv(m, names=['t','x','y'], delim_whitespace=True)
    f_1_0.append(_)
```

```python
fig, ax = plt.subplots(1,5, figsize=(20,4), sharex=True, sharey=True)

for i in range(4):
    ax[0].plot(f_1_0[i]['x'],f_1_0[i]['y'],lw=1)
    ax[0].set_aspect('equal')
    ax[0].minorticks_on()
    ax[0].tick_params(which='both',direction='in', top=True, right=True)
    ax[0].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[0].set_ylabel('Y ($\mu m$)', fontsize=16)
    ax[0].set_title('F = -1.0 pN', fontsize=16)
    
for i in range(4):
    ax[1].plot(f_0_5[i]['x'],f_0_5[i]['y'],lw=1)
    ax[1].set_aspect('equal')
    ax[1].minorticks_on()
    ax[1].tick_params(which='both', direction='in', top=True, right=True)
    ax[1].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[1].set_title('F = -0.5 pN', fontsize=16)
        
for i in range(4):
    ax[2].plot(f0_0[i]['x'],f0_0[i]['y'],lw=1)
    ax[2].set_aspect('equal')
    ax[2].minorticks_on()
    ax[2].tick_params(which='both', direction='in', top=True, right=True)
    ax[2].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[2].set_title('F = 0.0 pN', fontsize=16)
    
for i in range(4):
    ax[3].plot(f1_0[i]['x'],f1_0[i]['y'],lw=1)
    ax[3].set_aspect('equal')
    ax[3].minorticks_on()
    ax[3].tick_params(which='both', direction='in',top=True, right=True)
    ax[3].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[3].set_title('F = 1.0 pN', fontsize=16)
    
for i in range(4):
    ax[4].plot(f1_5[i]['x'],f1_5[i]['y'],lw=1)
    ax[4].set_aspect('equal')
    ax[4].minorticks_on()
    ax[4].tick_params(which='both', direction='in', top=True, right=True)
    ax[4].set_xlabel('X ($\mu m$)', fontsize=16)
    ax[4].set_title('F = 1.5 pN', fontsize=16)
    
#plt.savefig('fig/pulling_field.pdf', format='pdf', bbox_inches='tight')
```
