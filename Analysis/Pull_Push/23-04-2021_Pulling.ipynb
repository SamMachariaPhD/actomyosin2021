```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")
cm = 1/2.54
```

```python
pull = pd.read_csv('data/PullSpd_std77s5.csv', names=['spd','std'])
mtrs = pd.read_csv('data/PullBmtrs_std77s5.csv', names=['act_m','act_d','def_m','def_d'])
f = np.array([0.00,0.25,0.50,1.00,3.00,5.00,10.00,15.00])
```

```python
fig, ax = plt.subplots(3,1, figsize=(15*cm,30*cm), sharex=True)

ax[0].errorbar(f,pull['spd'],pull['std'], marker='.', markerfacecolor='none',lw=1,ls='--',capsize=3)
ax[0].minorticks_on()
ax[0].tick_params('both',direction='in', which='both')
#ax[0].set_xlabel('Pulling force (pN)', fontsize=14)
ax[0].set_ylabel('Speed ($\mu m/sec$)', fontsize=14)
#=====================================================
ax[1].errorbar(f,mtrs['act_m'],mtrs['act_d'], marker='.', markerfacecolor='lightgreen', lw=1,ls='--', capsize=3, color='green', label='Active myosin')
ax[1].errorbar(f,mtrs['def_m'],mtrs['def_d'], marker='.', markerfacecolor='lightblue', lw=1,ls='--', capsize=3, color='blue', label='Defective myosin')
ax[1].minorticks_on()
ax[1].tick_params('both', direction='in', which='both')
ax[1].set_ylabel('Binding motors', fontsize=14)
ax[1].legend()
#=====================================================
rto = mtrs['act_m']/(mtrs['act_m']+mtrs['def_m'])

ax[2].plot(f,rto, marker='.', markerfacecolor='none',lw=1,ls='--', color='darkgreen')
ax[2].minorticks_on()
ax[2].tick_params('both',direction='in', which='both')
ax[2].set_xlabel('Pulling force (pN)', fontsize=14)
ax[2].set_ylabel('Active ratio', fontsize=14)
#=====================================================
plt.tight_layout()
#plt.savefig('fig/pulling_and_binding.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
