```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
plt.style.use('ggplot')
cm = 1/2.54
```

```python
conf = glob.glob("data/conf/conf_t5R0.92F**.txt")
conf = sorted(conf)
```

```python
tip = []; beads = 13
for i in conf:
    _ = pd.read_csv(i, names=['t','x','y','z'], delim_whitespace=True)
    _ = _.iloc[0::beads]
    _ = _.reset_index(drop=True)
    tip.append(_)
```

```python
fig, ax = plt.subplots(4,5, figsize=(30*cm,20*cm))
f=0 # force index
ax[0,0].plot(tip[f]['t']/1e7,tip[f]['x'],lw=1)
ax[0,0].set_title('Pull = 0.0 pN')
ax[0,0].set_ylabel('X ($\mu m$)')
ax[1,0].plot(tip[f]['t']/1e7,tip[f]['y'],lw=1)
ax[1,0].set_ylabel('Y ($\mu m$)')
ax[2,0].plot(tip[f]['t']/1e7,tip[f]['z'],lw=1)
ax[2,0].set_ylabel('Z ($\mu m$)')
ax[2,0].set_xlabel('t (sec)')
ax[3,0].plot(tip[f]['x'],tip[f]['y'],lw=1)
ax[3,0].set_xlabel('X ($\mu m$)')
ax[3,0].set_ylabel('Y ($\mu m$)')
f=6
ax[0,1].plot(tip[f]['t']/1e7,tip[f]['x'],lw=1)
ax[0,1].set_title('Pull = 3.0 pN')
ax[1,1].plot(tip[f]['t']/1e7,tip[f]['y'],lw=1)
ax[2,1].plot(tip[f]['t']/1e7,tip[f]['z'],lw=1)
ax[2,1].set_xlabel('t (sec)')
ax[3,1].plot(tip[f]['x'],tip[f]['y'],lw=1)
ax[3,1].set_xlabel('X ($\mu m$)')
f=7
ax[0,2].plot(tip[f]['t']/1e7,tip[f]['x'],lw=1)
ax[0,2].set_title('Pull = 5.0 pN')
ax[1,2].plot(tip[f]['t']/1e7,tip[f]['y'],lw=1)
ax[2,2].plot(tip[f]['t']/1e7,tip[f]['z'],lw=1)
ax[2,2].set_xlabel('t (sec)')
ax[3,2].plot(tip[f]['x'],tip[f]['y'],lw=1)
ax[3,2].set_xlabel('X ($\mu m$)')
f=4
ax[0,3].plot(tip[f]['t']/1e7,tip[f]['x'],lw=1)
ax[0,3].set_title('Pull = 10.0 pN')
ax[1,3].plot(tip[f]['t']/1e7,tip[f]['y'],lw=1)
ax[2,3].plot(tip[f]['t']/1e7,tip[f]['z'],lw=1)
ax[2,3].set_xlabel('t (sec)')
ax[3,3].plot(tip[f]['x'],tip[f]['y'],lw=1)
ax[3,3].set_xlabel('X ($\mu m$)')
f=5
ax[0,4].plot(tip[f]['t']/1e7,tip[f]['x'],lw=1)
ax[0,4].set_title('Pull = 15.0 pN')
ax[1,4].plot(tip[f]['t']/1e7,tip[f]['y'],lw=1)
ax[2,4].plot(tip[f]['t']/1e7,tip[f]['z'],lw=1)
ax[2,4].set_xlabel('t (sec)')
ax[3,4].plot(tip[f]['x'],tip[f]['y'],lw=1)
ax[3,4].set_xlabel('X ($\mu m$)')

plt.tight_layout()
#plt.savefig('fig/tip_behaviour_v2.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
