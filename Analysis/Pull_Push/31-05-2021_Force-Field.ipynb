```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
cm = 1/2.54
```

```python
dta = pd.read_csv('data/spd_std77.csv', names=['spd','std'])
F = np.array([5.0,4.5,4.0,1.5,1.0,0.5,0.0,-7.5,-5.0,-2.5])

zalinge = pd.read_csv('data/zalinge.csv')
zalinge_hi = pd.read_csv('data/zalinge-hi.csv')
zalinge_lo = pd.read_csv('data/zalinge-lo.csv')
zalinge_err = zalinge_hi['y'] - zalinge_lo['y']
```

```python
zalinge_err
```

```python
fig, ax = plt.subplots(1,1,figsize=(10*cm,10*cm))

ax.errorbar(F,dta['spd'],dta['std'], marker='o', lw=1, ls='--', c='brown', markerfacecolor='lime', capsize=3)
ax.set_xlabel('Force (pN)', fontsize=16)
ax.set_ylabel('Speed ($\mu m$)', fontsize=16)

#plt.savefig('fig/force-speed.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10*cm,10*cm))

#ax.plot(zalinge['x'],zalinge['y'], marker='o', lw=1, ls='--', c='brown', markerfacecolor='lime')
ax.errorbar(zalinge['x'],zalinge['y'], yerr=zalinge_err-2.7, marker='o', lw=1, ls='--', c='brown', markerfacecolor='lime', capsize=3)

ax.set_xlabel('Field (V/m)', fontsize=16)
ax.set_ylabel('Mean velocity ($\mu m/s$)', fontsize=16)

ax.minorticks_on()
ax.tick_params('both',direction='in',top=False,right=False,length=8, which='major', labelsize=14)
ax.tick_params('both',direction='in',top=False,right=False,length=4, which='minor', labelsize=14)

ax.set_xticks(np.arange(0,8001,2000))
ax.set_yticks(np.arange(0,11,2))
ax.tick_params(direction='in', which='both')

#plt.savefig('fig/zalinge.png', format='png', dpi=300, bbox_inches='tight')
plt.show()
```
