```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
cm = 1/2.54
```

```python
ac_def_27_0042 =pd.read_csv('data0042/mtrs_s27s5kBT0042.csv', names=['spec1','dev1','spec2','dev2'])
BM_r_27_0042 =pd.read_csv('data0042/BM_ratio_std27s5kBT0042.csv', names=['spec1','dev1','spec2','dev2'])
r_27_0042 = np.array([0.70,0.80,0.82,0.84,0.86,0.88,0.90,0.92,0.94,0.96,0.98,1.00])
s_77_0042 = pd.read_csv('data0042/v11nspd_std27t5kBT0042.csv', names=['spd','std']) 
```

```python
fig, ax = plt.subplots(1,1, figsize=(10*cm,10*cm), sharex=False)


ax.errorbar(r_27_0042,ac_def_27_0042['spec1'],ac_def_27_0042['dev1'], capsize=3, linestyle='--', linewidth=1, elinewidth=1, color='green', marker='o', markerfacecolor='lightgreen', ecolor='green', label='Active')
ax.errorbar(r_27_0042,ac_def_27_0042['spec2'],ac_def_27_0042['dev2'], capsize=3, linestyle='--', linewidth=1, elinewidth=1, color='blue', marker='o', markerfacecolor='lightblue', ecolor='blue', label='Defective')


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.minorticks_on()
ax.tick_params('both',direction='in',top=False,right=False,length=8,which='major', labelsize=14)
ax.tick_params('both',direction='in',top=False,right=False,length=4,which='minor', labelsize=14)

ax.set_xlabel('$r_{substrate}$', fontsize=16)
#ax.set_ylabel(r'$\overline{n}_{binding}$', fontsize=16) # not standard error of mean
ax.set_ylabel(r'$n_{binding}$', fontsize=16)
#================================================================================

# plt.savefig('fig/BM_rto_v4-1-v2.pdf', format='pdf', dpi=500, bbox_inches='tight')
# plt.savefig('fig/BM_rto_v4-1-v2.svg', format='svg', dpi=500, bbox_inches='tight')
# plt.savefig('fig/BM_rto_v4-1-v2.png', format='png', dpi=500, bbox_inches='tight')

plt.show()
```
