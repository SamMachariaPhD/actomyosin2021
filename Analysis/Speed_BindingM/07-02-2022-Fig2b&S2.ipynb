```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
cm = 1/2.54
```

```python
s_77_0042 = pd.read_csv('data0042/v11nspd_std77t5kBT0042.csv', names=['spd','std']) 
r_77_0042 = np.array([0.82,0.84,0.86,0.88,0.90,0.92,0.94,0.96,0.98,1.00])
r_27_0042 = np.array([0.70,0.80,0.82,0.84,0.86,0.88,0.90,0.92,0.94,0.96,0.98,1.00])
s_27_0042 = pd.read_csv('data0042/v11nspd_std27t5kBT0042.csv', names=['spd','std']) 

s_77_0042MD2000 = pd.read_csv('data0042/v11nspd_std77t5kBT0042MD2000.csv', names=['spd','std']) 
s_77_0042ATP500 = pd.read_csv('data0042/v11nspd_std77t5kBT0042ATP500.csv', names=['spd','std']) 
s_77_0042AF2um = pd.read_csv('data0042/v11nspd_std77t5kBT0042AF2um.csv', names=['spd','std']) 
```

```python
fig, ax = plt.subplots(1,1, figsize=(10*cm,10*cm), sharex=True, sharey=True)

ax.errorbar(r_27_0042,s_27_0042['spd'],s_27_0042['std'], capsize=3, linestyle='--', lw=1, color = 'red', mec = 'red', marker='o',\
             markerfacecolor='lightsalmon', ecolor='red', elinewidth=1, label='v11a')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
#ax.spines['left'].set_bounds(0,10)

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.minorticks_on()
ax.tick_params('both',direction='in',top=False,right=False,which='major', labelsize=14) # ,length=8, width=0.5
ax.tick_params('both',direction='in',top=False,right=False,which='minor', labelsize=14)

ax.set_xlabel('$r_{substrate}$', fontsize=18)
ax.set_ylabel('Speed ($\mu m/s$)', fontsize= 18)

#================================================================

# plt.savefig('fig/speed_v3.pdf', format='pdf', dpi=500, bbox_inches='tight')
# plt.savefig('fig/speed_v3.svg', format='svg', dpi=500, bbox_inches='tight')
# plt.savefig('fig/speed_v3.png', format='png', dpi=500, bbox_inches='tight')

plt.show()
```
