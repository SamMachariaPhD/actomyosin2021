```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

cm = 1/2.54
```

```python
# https://automeris.io/WebPlotDigitizer/
# https://apps.automeris.io/wpd/
mr = pd.read_csv('data_22-06/web_plot_digitizer.csv', names=['mr','spd']) # molar ratio
```

$r = \frac{HMM}{HMM+LMM}$ <br>
$r = \frac{1}{1+\frac{LMM}{HMM}}$

```python
mr_r = (1/(1+(mr['mr'])))
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,5), sharey=True)

ax[0].plot(mr['mr'],mr['spd'], color='brown', marker ='o', markerfacecolor='lightgrey', ls='--', lw=1)
ax[0].set_xlabel('$LMM/HMM$ (molar ratio)', fontsize=14)
ax[0].set_ylabel('Mean Velocity ($\mu m/sec$)', fontsize=14)
ax[0].set_yticks(np.arange(2,13,2))
ax[0].set_ylim(2,12)
ax[0].minorticks_on()
ax[0].tick_params(which='both', direction='in')

ax[1].plot(mr_r,mr['spd'], color='brown', marker ='o', markerfacecolor='lightgrey', ls='--', lw=1)
ax[1].set_xlabel(r'$R = \frac{1}{1+\frac{LMM}{HMM}}$', fontsize=14)
ax[1].minorticks_on()
ax[1].tick_params(which='both', direction='in')
#ax[1].set_xticks(np.arange(0.08,0.181,0.02))

#plt.savefig('fig/hmm_lmm.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
