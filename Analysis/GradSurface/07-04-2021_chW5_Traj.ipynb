```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import glob
import random

cm = 1/2.54
```

```python
traj_ = glob.glob('data/chW5/Tip**.txt')
traj_ = sorted(traj_)

actM_ = glob.glob('data/chW5/bmAct**.csv')
actM_ = sorted(actM_)

traj2_ = glob.glob('data/chW2/Tip**.txt')
traj2_ = sorted(traj2_)

actM2_ = glob.glob('data/chW2/bmAct**.csv')
actM2_ = sorted(actM2_)
```

```python
traj = []; traj2 = []; actM = []; actM2 = []
for i in traj_:
    traj.append(pd.read_csv(i,names=['t','x','y'], delim_whitespace=True))

for j in traj2_:
    traj2.append(pd.read_csv(j,names=['t','x','y'], delim_whitespace=True))
    
for k in actM_:
    actM.append(pd.read_csv(k,names=['act'], delim_whitespace=True))
    
for l in actM2_:
    actM2.append(pd.read_csv(l,names=['act'], delim_whitespace=True))
```

```python
trajAct = []; trajAct2 = []
for i in range(len(traj)):
    trajAct.append(pd.concat([traj[i],actM[i]],axis=1))
    
for j in range(len(traj2)):
    trajAct2.append(pd.concat([traj2[j],actM2[j]],axis=1))
```

```python
idx = []; idx2 = []
for i in range(len(traj)):
    idx.append(trajAct[i][trajAct[i].act == 0].index.values) # the index where "act = 0"
    
for j in range(len(traj2)):
    idx2.append(trajAct2[j][trajAct2[j].act == 0].index.values)    
```

```python
tjAct = []; tjAct2 = []
for i in range(len(traj)):
    print("i = %s"%i)
    try:
        tjAct.append(trajAct[i].drop(index=np.arange(idx[i][0].tolist(),traj[i].shape[0],1).tolist())) # discard rows starting from the index where "act = 0" to the end
    except:
        print("blank_i = %s"%i)
        tjAct.append(trajAct[i])
    
for j in range(len(traj2)):
    print("j = %s"%j)
    try:
        tjAct2.append(trajAct2[j].drop(index=np.arange(idx2[j][0].tolist(),traj2[2].shape[0],1).tolist()))
    except:
        print("blank_j = %s"%j)
        tjAct2.append(trajAct2[j])
```

```python
fig,ax = plt.subplots(1,2,figsize=(30*cm,15*cm), sharex=True, sharey=False)

for i in range(len(traj)):
    ax[0].plot(tjAct[i]['x'],tjAct[i]['y'], lw=0.5)#, marker='.', markersize=2, markerfacecolor='none', ls='--')

##ax.plot(traj[0]['x'],traj[0]['y'], lw=1)
#ax[0].plot(xRnd_,yRnd_,marker='.', markersize=2,ls='none',color='green',alpha=0.3,label='Myosin')

ax[0].minorticks_on()
ax[0].tick_params('both',direction='in',top=True,right=True,bottom=True,left=True,length=8, width=0.5,which='major', labelsize=14)
ax[0].tick_params('both',direction='in',top=True,right=True,bottom=True,left=True,length=4, width=0.4,which='minor', labelsize=14)

ax[0].set_aspect('equal','box')
ax[0].set_xticks(np.arange(2,13,1))
ax[0].set_yticks(np.arange(-2.5,2.6,1))

ax[0].set_xlabel('X($\mu m$)', fontsize=15)
ax[0].set_ylabel('Y($\mu m$)', fontsize=15)

for j in range(len(traj2)):
    ax[1].plot(tjAct2[j]['x'],tjAct2[j]['y'], lw=0.5)#, marker='.', markersize=2, markerfacecolor='none', ls='--')

#ax[1].plot(xRnd2_,yRnd2_,marker='.', markersize=2,ls='none',color='green',alpha=0.3,label='Myosin')

ax[1].minorticks_on()
ax[1].tick_params('both',direction='in',top=True,right=True,bottom=True,left=True,length=8, width=0.5,which='major', labelsize=14)
ax[1].tick_params('both',direction='in',top=True,right=True,bottom=True,left=True,length=4, width=0.4,which='minor', labelsize=14)

ax[1].set_aspect('equal','box')
ax[1].set_xticks(np.arange(3,11,1))
ax[1].set_yticks(np.arange(-1,1.1,1))

ax[1].set_xlabel('X($\mu m$)', fontsize=15)
##ax[1].set_ylabel('Y($\mu m$)', fontsize=15)
plt.text(-2.7,-4,'(a)',ha='center', fontsize=14)
plt.text(6.5,-2.3,'(b)', ha='center', fontsize=14)

#plt.savefig('fig/gradSurface.pdf', format='pdf', dpi=500, bbox_inches='tight')

plt.show()
```
