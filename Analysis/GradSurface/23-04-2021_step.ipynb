```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
plt.style.use("ggplot")
cm = 1/2.54
```

```python
stp0 = glob.glob('data/step/TipXY0**.txt')
stp0 = sorted(stp0)
stp1500 = glob.glob('data/step/TipXY1500**.txt')
stp1500 = sorted(stp1500)
stp3000 = glob.glob('data/step/TipXY3000**.txt')
stp3000 = sorted(stp3000)
```

```python
md0 = []; md1500 = []; md3000 = []

for i in stp0:
    _ = pd.read_csv(i, names=['t','x','y'], delim_whitespace=True)
    md0.append(_)
    
for j in stp1500:
    _ = pd.read_csv(j, names=['t','x','y'], delim_whitespace=True)
    md1500.append(_)
    
for k in stp3000:
    _ = pd.read_csv(k, names=['t','x','y'], delim_whitespace=True)
    md3000.append(_)
```

```python
fig, ax = plt.subplots(1,3, figsize=(30*cm,100*cm), sharex=True, sharey=True)


for l in range(len(md0)):
    ax[0].plot(md0[l]['x'],md0[l]['y'])
ax[0].set_aspect('equal','box')
ax[0].set_title(r'$\rho = (0_{up}, 3000_{down})$')
    
for m in range(len(md1500)):
    ax[1].plot(md1500[m]['x'],md1500[m]['y'])
ax[1].set_aspect('equal','box')
ax[1].set_title(r'$\rho = (1500_{up}, 3000_{down})$')

for n in range(len(md3000)):
    ax[2].plot(md3000[n]['x'],md3000[n]['y'])
ax[2].set_aspect('equal','box')
ax[2].set_title(r'$\rho = (3000_{up}, 3000_{down})$')

plt.tight_layout()
plt.show()
```
